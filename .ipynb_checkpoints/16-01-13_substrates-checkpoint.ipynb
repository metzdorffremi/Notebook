```python
samplelist = H+S

def collect_param(samples, param, quality=0):
    l = []
    for sample in samples:
        if sample["quality"] >= quality:
            l.append(sample[param])
    return l
```

```python
names = collect_param(samplelist, 'name')


scale = 2.
fig = plt.figure(1, figsize=(8*scale,6*scale))
plt.clf()

ax1 = plt.subplot2grid((3,2), (0,0), rowspan=2)
ax2 = plt.subplot2grid((3,2), (0,1))
ax3 = plt.subplot2grid((3,2), (1,1))
ax4 = plt.subplot2grid((3,2), (2,0), colspan=2)

RoCas = collect_param(samplelist, 'RoCa', 0)
RoCbs = collect_param(samplelist, 'RoCb', 0)

ax1.hist((np.array(RoCbs) + np.array(RoCas))*1e3/2., bins=10, color="blue", alpha=0.5, range=(0.5,2.5))
ax1.set_xlabel("$\mathrm{RoC\ mean\ [mm]}$")
ax2.hist(np.array(RoCas)*1e3, bins=10, color="green", alpha=0.5, range=(0.5,2.5))
ax2.set_xlabel("$\mathrm{RoCa\ [mm]}$")
ax3.hist(np.array(RoCbs)*1e3, bins=10, color="red", alpha=0.5, range=(0.5,2.5))
ax3.set_xlabel("$\mathrm{RoCb\ [mm]}$")
ax4.hist(np.array(RoCbs)/np.array(RoCas), bins=7, color="yellow", alpha=0.5, range=(0.7,1.))
ax4.set_xlabel("$\mathrm{RoCb/RoCa}$")
```
