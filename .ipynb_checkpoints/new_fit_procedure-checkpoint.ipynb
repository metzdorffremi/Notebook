```python
import scipy.optimize as optimization

LAMBDA = 1066.e-9
CONSTANT_c = 299792458
consider_gouy = False
params_gouy = [-6.21366e-7, 0.0153648]
R0 = 1300.e-6

##########collect the relevant parmaters for the analysis for each peak##########
collect_params = True
if collect_params:
    pks  = [c.pk for c in parentCurve.childs.all() if c.pk > 0 and c.pk <= last_fitted]
    
    peak_rel_width = []
    peak_sb_width = []
    peak_phases = []
    steps = []
    steps_single = []
    xpzts = []
    ramp_scales = []
    ramp_fs = []
    ramp_phis = [] 
    parts = [0]
    measured_length = []
    fit_rmss = []

    for i, pk in enumerate(pks):                                                #boucle sur tous les scans
        fsr = CurveDB.objects.get(pk=pk)
        ramp = fsr.childs.filter(_name__contains = 'ramp')[0]
        
        if not "ramp_phi" in fsr.params:
            fit = ramp.fit("sin",autosave=False)
            fp = fit.params
            fsr.params.update(dict(ramp_f=fp["f"],
                                   ramp_scale=fp["scale"],
                                   ramp_phi=fp["phi"]))
            fsr.save()
        stepper_pos = fsr.params["stepper_pos"]
        ramp_f = fsr.params["ramp_f"]
        ramp_scale  = fsr.params["ramp_scale"]
        ramp_phi = fsr.params["ramp_phi"]
        sb_freq = fsr.params["sb_freq"]
        measured_length.append(fsr.params["length"])

        if ramp_scale >0:
            ramp_phi += np.pi
            ramp_scale *= -1.
        if "sweep_direction" in fsr.params:
            direction = fsr.params["sweep_direction"]
        else:
            direction = "back"

        if stepper_pos %1 == 0 and direction == "back":                         #ne prend que les scans fait en reculant le miroir
            steps.append([])
            steps_single.append(step_max - stepper_pos)
            ramp_scales.append([])
            xpzts.append([])
            peak_phases.append([])
            ramp_fs.append([])
            ramp_phis.append([])
            peak_rel_width.append([])
            peak_sb_width.append([])
            fit_rmss.append([])
            
            parts[-1] += 1

            for peak in fsr.childs.filter(_name__contains="peak@"):             #boucle sur les pics de chaque scan
                #if peak.params["slope"] == 1:
                fit_peak = peak.childs.first()
                t = fit_peak.params["x0"]
                fit_rmss[-1].append(fit_peak.params["fit_rms"])
                peak_rel_width[-1].append(fit_peak.params["relative_bandwidth"])
                peak_sb_width[-1].append(fit_peak.params["SBwidth"])                
                peak_phases[-1].append(2*np.pi*ramp_f*t - ramp_phi)
                steps[-1].append(step_max - stepper_pos)
                ramp_scales[-1].append(ramp_scale)
                ramp_fs[-1].append(ramp_f)
                ramp_phis[-1].append(ramp_phi)
        elif direction == "forth" and parts[-1]<>0:
            parts.append(0)
    
    print "Done"
N = len(steps)

##########my functions##########
def calc_losses(length, bandwidth):
    kappa_over_2pi = np.abs(2*bandwidth)
    finesse = CONSTANT_c / (2*length*kappa_over_2pi)
    losses = 2*np.pi / finesse
    return losses

def calc_length(params, step):
    slope, y0 = params
    length = slope * (step_max - step) + y0
    return length

def delta_fsr(params, step):
    if not consider_gouy:
        return 1.
    length = calc_length(params, step)
    sq_waist = LAMBDA / np.pi * np.sqrt(length*(R0 - length))
    zr = np.pi * sq_waist / LAMBDA
    d_phig = (1./zr) / (1 + (length/zr)**2)
    k = 2*np.pi / LAMBDA
    length_fsr = k / (k - d_phig)
    return length_fsr
print delta_fsr(params_gouy, 0), delta_fsr(params_gouy, 1000) 

def pzt_pos(params, scale, phase):
    """
    return the pzt position for a peak located by its phase (2*pi*f*t)
    the position is estimated by a polynomial function of the ramp sin
    """
    a0, a1, a2, a3, a4, phaselag1, phaselag2, phaselag3, phaselag4 = params
    voltage1 = scale * np.sin(phase + phaselag1)
    voltage2 = scale * np.sin(phase + phaselag2)
    voltage3 = scale * np.sin(phase + phaselag3)
    voltage4 = scale * np.sin(phase + phaselag4)
    return a0 + a1 * voltage1 + a2 * voltage2**2 + a3 * voltage3**3 + a4 * voltage4**4

def pzt_speed(params, scale, phase, f = 4.):
    a0, a1, a2, a3, a4, phaselag1, phaselag2, phaselag3, phaselag4 = params
    speed1 = scale * 2*np.pi * 1. * f * np.cos(phase + phaselag1)
    speed2 = scale * 2*np.pi * 2. * f * np.cos(phase + phaselag2) * np.sin(phase + phaselag2)
    speed3 = scale * 2*np.pi * 3. * f * np.cos(phase + phaselag3) * np.sin(phase + phaselag3)**2
    speed4 = scale * 2*np.pi * 4. * f * np.cos(phase + phaselag4) * np.sin(phase + phaselag4)**3
    return a1 * speed1 + a2 * speed2 + a3 * speed3 + a4 * speed4

def sb_width(params, scale, phase, f, length, sb_freq = sb_freq):
    return abs(sb_freq * length / pzt_speed(params, scale, phase, f) / (CONSTANT_c / LAMBDA))

def guess_offset(params, scale, phases, step, threshold = 0.5):
    xpzts = pzt_pos(params, scale, phases)
    min_offset = min(xpzts)
    first_peak = []
    for xpzt in xpzts:
        if xpzt - min_offset < threshold:
            first_peak.append(xpzt)
    return mean((xpzts - mean(first_peak) +0.5) % delta_fsr(params_gouy, step) - 0.5 + mean(first_peak))

def delta_pzt(params, scale, phases, step):
    """
    calculate the gap between the pzt position given by the fit and the expected position based on the FSR
    """
    offset = guess_offset(params, scale, phases, step)
    xpzts = pzt_pos(params, scale, phases)
    return ((xpzts - offset) + 0.5) % delta_fsr(params_gouy, step) - 0.5

def opt_func(params, ramp_scales, peak_phases, steps, opt_steps = [], start = 0, stop = N):
    opt_deltas = []
    for i in range(start, stop):
        deltas = delta_pzt(params, array(ramp_scales[i]), array(peak_phases[i]), steps[i][0])
        for delta, step in zip(deltas, steps[i]):
            opt_steps.append(step)
            opt_deltas.append(delta)
    return opt_deltas

##########Initial parameters##########
phaselag = -24*np.pi/180.
params0 = array([2,0.6,0.,0.,0.,phaselag,0.,0.,0.])

##########Find fit params and plot deltas before and after##########

figure(1)
opt_steps = []
opt_deltas0 = opt_func(params0, ramp_scales, peak_phases, steps, opt_steps=opt_steps)
plot(opt_deltas0, opt_steps, '+', label=u"$\mathrm{pre\ opt}$")
opt_steps = []
t = time.time()
params_fit =  optimization.leastsq(opt_func, params0, args=(ramp_scales, peak_phases, steps, []))[0]
print time.time() - t
opt_deltas = opt_func(params_fit, ramp_scales, peak_phases, steps, opt_steps=opt_steps)
plot(opt_deltas, opt_steps, 'x', label=u"$\mathrm{post\ opt}$")
legend()


##########Just to check the waveform##########
"""
opt_phases, opt_scales = [], []
for phases, scales in zip(peak_phases, ramp_scales):
    for phase, scale in zip(phases, scales):
        opt_phases.append(phase)
        opt_scales.append(scale)
plot(sin(array(opt_phases)+-24*np.pi/180.), pzt_pos(params_fit, array(opt_scales), array(opt_phases)), '+')
"""

##########Determine if a curve is ok or not##########
"""
deltas = []
nicecurves = []
for i in range(N):
    deltas.append(delta_pzt(params_fit, ramp_scales[i], peak_phases[i], steps[i][0]))
    if std(deltas[-1]) > 0.05:
        nicecurves.append(False)
    else:
        nicecurves.append(True)


##########Fill the list offsets##########

offsets = []
for i in range(N):
    offsets.append(guess_offset(params_fit, ramp_scales[i], peak_phases[i], steps[i][0]))
    if nicecurves[i]:
        if offsets[i] > 1.:
            nicecurves[i] = False
"""

##########Compute the stepsize##########
"""
stepsizes = array(offsets[1:])-array(offsets[:-1])
plot(steps_single[:-1],stepsizes)
for i in range(N-1):
    if stepsizes[i] < 0.:
        stepsizes[i] += 1.  
    if stepsizes[i] >.5 and stepsizes[i] <1.:
        stepsizes[i] -=1.
    if abs(stepsizes[i] - mean(stepsizes)) > 0.5:
        stepsizes[i] = stepsizes[i]#mean(stepsizes)
stepsizes += 1.
plot(steps_single[:-1],stepsizes)
"""

##########New sort of data and plot stepsize##########
"""
x, y=[], []
mean_stepsize = mean(stepsizes)
for i in range(N-1):
    if abs(stepsizes[i] - mean_stepsize) > .5:
        nicecurves[i] = False
    else:
        x.append(steps_single[i])
        y.append(stepsizes[i])
x = step_max - array(x)
y = (array(y) * LAMBDA/2.) * 1e6
plot(x, y, label=u"$\mathrm{experiment}$")
plot([x[0], x[-1]], [mean(y), mean(y)], label=u"$\mathrm{mean=%.3f\ \mu m}$"%(mean(y)))
grid('on')
xlabel(u"$\mathrm{stepper\ position}$")
ylabel(u"$\mathrm{step\ size\ [\mu m]}$")
title(u"$\mathrm{Motorized\ translation\ stage\ 8MT173\ (STANDA)}$")
legend(loc="lower left")
#savefig("stepsize_vs_step_STANDA_8MT173.pdf", bbox_inches='tight')
"""

##########Plot losses_vs_length##########
"""
xtot = []
ytot = []
lengths = []
for i in range(N-1):
    length = 140.1e-6/(LAMBDA/2.) + 12 + sum(stepsizes[:i]) - array(pzt_pos(params_fit, ramp_scales[i], peak_phases[i]))
    lengths.append(length)
    if nicecurves[i]:
        #plot(steps[i], length, '+')
        #length = sum(stepsizes[:i])+2.*140.e-6/LAMBDA + array(pzt_pos(params_fit, ramp_scales[i], peak_phases[i])) - array(deltas[i])
        #plot(length, 1e6 * calc_losses(length * LAMBDA/2., sb_freq * array(peak_rel_width[i])), ".", color='green')
        semilogy()
        ytot+=(list(calc_losses(length * LAMBDA/2., sb_freq * array(peak_rel_width[i]))))
        xtot+=(list(length))

modenumbers = []
losses = []
for x in xtot:
    modenumbers.append(int(x))
clear_modenumbers = []
clear_losses = []
for n in range(min(modenumbers), max(modenumbers)+1):
    temp_losses = []
    for modenumber, losses in zip(modenumbers, ytot):
        if modenumber == n:
            temp_losses.append(losses)
    clear_modenumbers.append(n)
    clear_losses.append(mean(temp_losses))
plot(clear_modenumbers, array(clear_losses)*1e6)
xlabel(u"$\mathrm{mode\ number}$")
ylabel(u"$\mathrm{losses\ [ppm]}$")
plt.xlim(min(clear_modenumbers), max(clear_modenumbers))
title(u"$\mathrm{run\ %g}$"%(parent_pk))
grid('on')
semilogy()   
#savefig("losses_vs_modenumber.pdf")
#curve = CurveDB.create(array(clear_modenumbers) * LAMBDA/2., clear_losses)
#curve.name = "T+P vs length new analysis %g"%(parent_pk)
#curve.save()
"""

##########fit_rms_vs_step#########
"""
x, y1, y2, y21, y3, y4 = [], [], [], [], [], []
print len(lengths), len(fit_rmss)
for i in range(N-1):
    if nicecurves[i]:
        for j in range(len(steps[i])):
            if fit_rmss[i][j] < 1.005:
                x.append(step_max - steps[i][j])
                y1.append(fit_rmss[i][j])
                y21.append(calc_losses(lengths[i][j] * LAMBDA/2., sb_freq * peak_rel_width[i][j] * peak_sb_width[i][j] /
                                       sb_width(params_fit, ramp_scales[i][j], peak_phases[i][j], 4, lengths[i][j], sb_freq = sb_freq)))
                y2.append(calc_losses(lengths[i][j] * LAMBDA/2., sb_freq * peak_rel_width[i][j]))
                y3.append(array(peak_sb_width[i][j]))
                y4.append(sb_width(params_fit, ramp_scales[i][j], peak_phases[i][j], 4, lengths[i][j], sb_freq = sb_freq))
                #plot( step_max - array(steps[i]), array(fit_rmss[i]), '.', color = "green")
                #plot( step_max - array(steps[i]) , calc_losses(lengths[i] * LAMBDA/2., sb_freq * array(peak_rel_width[i])), '+', color='blue')
                #plot( step_max - array(steps[i]) , array(peak_sb_width[i]), '.', color='red')
plot(x, y1, '.', color = "green")
plot(x, y21, '.', color = 'blue')
plot(x, y2, '+', color='red')
#plot(x, y3, '.', color='red')
plot(x, array(y4)/array(y3), '.', color='orange')
grid('on')
semilogy()             
"""
"""
for i in range(N):
    if nicecurves[i]:
        #plot( pzt_pos(params0, ramp_scales[i], peak_phases[i])
        #     -guess_offset(params0, ramp_scales[i], peak_phases[i], steps[i][0]),
        #     steps[i], '+', color='blue')
        plot(pzt_pos(params_fit, ramp_scales[i], peak_phases[i])
             -guess_offset(params_fit, ramp_scales[i], peak_phases[i], steps[i][0]),
             steps[i], '+', color= "green"
"""
```
